```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/book.csv')
df.head()

antiguos = df[df['original_publication_year'] < 2000].shape[0]
recientes = df[df['original_publication_year'] >= 2000].shape[0]
#crear un ciclo que calcule el promedio de todos los aos con un acumulador
suma = 0
cantidad = 0

for year in df['original_publication_year']:
    if pd.notna(year):
        suma += year
        cantidad += 1
print(round(suma / cantidad))
# Calcular total y porcentajes
promedio= (antiguos + recientes) / 2
total = antiguos + recientes

porcentaje_antiguos = antiguos / total * 100
porcentaje_recientes = recientes / total * 100

# Graficar en porcentaje
plt.figure(figsize=(5, 6))
plt.bar(['Recientes', 'Antiguos'], [porcentaje_recientes, porcentaje_antiguos], color=['purple', 'blue'])
plt.title('Distribucin de libros (porcentaje)')
plt.ylabel('Porcentaje de libros')
plt.ylim(0, 100)  # Eje Y de 0% a 100%
plt.show()

autores_filtrados = ['Nicholas Sparks', 'Stephenie Meyer', 'Suzanne Collins', 'Jane Austen']
df_filtrado = df[df['authors'].isin(autores_filtrados)]
conteo = df_filtrado['authors'].value_counts()

plt.figure(figsize=(5, 4))
conteo.plot(kind='bar', color='teal')
plt.title('Cantidad de libros publicados por autor')
plt.ylabel('Cantidad de libros')
plt.xlabel('Autor')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
